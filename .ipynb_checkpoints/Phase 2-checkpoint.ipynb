```python
import numpy as np
import pandas as pd
```

```python
# import dataset
crimes = pd.read_csv("Dirty_Data/crime.csv", encoding='latin-1', engine='python')
```

```python
crimes.head()
```

```python
#extract the date from first occurance 
crimes['occurrence_date'] = crimes['FIRST_OCCURRENCE_DATE'].str.extract('(\d+/\d+/\d+)', expand=True)
```

```python
crimes.head()
```

```python
crimes['occurence_times'] = crimes['FIRST_OCCURRENCE_DATE'].str.extract('((?:1[0-2]|0?[0-9]):[0-5][0-9]:[0-5][0-9] ([AaPp][Mm]))', expand = True)
```
